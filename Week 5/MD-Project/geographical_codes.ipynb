```python
from haversine import haversine
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Importing the data set

```python
df=pd.read_excel("Data/Data_MidTerm_Project_Real_State_Regression.xls").reset_index()
```

#### Creating lat_long column in df

```python

df["lat_long"] = list(zip(df["lat"], df["long"]))
```

#### Finding centre point of seattle and bellevue

```python
seattle=[47.6092,-122.3363]
bellevue= [47.61555,-122.20392]
```

```python
seattle_distances= []
for i in df["lat_long"]:
    seattle_distances.append(haversine((seattle),(i)))
df["distance_from_seattle"]= pd.Series(seattle_distances)

bellevue_distances= []
for i in df["lat_long"]:
    bellevue_distances.append(haversine((bellevue),(i)))
df["distance_from_bellevue"]= pd.Series(bellevue_distances)
```

#### Testing the results

```python
bellevue= [47.61555,-122.20392]
seattle=[47.6092,-122.3363]
test= [47.5941, -122.299] # first row of the df
test1= haversine(seattle,test)
test2= haversine(bellevue,test)
test1,test2

### There is an error of close to 500 m in the calculations 
```
