## we need to standarize the heading names

making the column names lower case #vamos a poner el nombre de las columnas en minusculas

```python
def lower_case_column_names(hk_df):
    hk_df.columns=[i.lower() for i in hk_df.columns]
    return hk_df
```

```python
# functión para unificar todas las categorias de la columna gender. 
#Como resultado tendremos solo dos posibles resultados: male o female
def clean_gender(x):
    if x in ['M', 'MALE']:
        return 'Male'
    elif x.startswith('F'):
        return 'Female'
    else:
        return 'U'
```

```python
list(map(clean_gender, hk_df['gender']))
```
