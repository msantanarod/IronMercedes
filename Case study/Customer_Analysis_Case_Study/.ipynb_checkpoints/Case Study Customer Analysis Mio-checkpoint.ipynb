## Load data.


```python
file1=pd.read_csv("Data/file1.csv")
file2=pd.read_csv("Data/file2.csv")
file3=pd.read_csv("Data/file3.csv")
```

## Standardizing header names

We can see that some columns name are in capital letter and another one in small letter. 

```python
ca_df.columns
# This function allows us to see the index
# We can see that the "Gender" column name is in capital letter and the rest of column names are in small letter
```

```python
def lower_case_column_names(ca_df):
    ca_df.columns=[i.lower() for i in ca_df.columns]
    return ca_df
```

```python
# This function allows us to see all column names in small letter
lower_case_column_names(ca_df)
```

```python
# wtih function rename i will change the columns name. 
#from "customer lifetime value to customer_lifetime_value"
def rename_columns(ca_df):
    ca_df.rename(columns={'customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto', 'number of open complaints':'number_of_open_complaints',"policy type":"policy_type", "vehicle class":"vehicle_class","total claim amount": "total_claim_amount"}, inplace=True )
    return ca_df
```

```python
rename_columns(ca_df)
```

```python
def drop_columns(ca_df):
    ca_df.drop(columns=["customer"], inplace=True)
    return ca_df
```

## Working with Data type

```python
# checking data type of our data set
```

```python
ca_df.info()
# We this function we can see the data type.
#Variables customer_lifetime_value and number of total complaints are not correct
#they should not be objects
```

```python
# We can use the function to_numeric to transform the data into a number(integer or float)

ca_df["customer_lifetime_value"] = pd.to_numeric(ca_df["customer_lifetime_value"], errors='coerce') 

ca_df["number_of_open_complaints"] = pd.to_numeric(ca_df["number_of_open_complaints"], errors='coerce')
```

```python
ca_df.info()
```

```python
# variable "number_of_open_complaints" should not be a float number. 
# We can see that there is many NaN values in the columns.
#We cannot transform "number_of_open_complaints" into an integer because of Nan values 
#In this case, we have to replace the NaN values with "0" with replace function. 
ca_df['number_of_open_complaints'] = ca_df['number_of_open_complaints'].fillna(0)
ca_df['number_of_open_complaints']

```

```python
# we can transform "number_of_open_complaints" into an integer
ca_df['number_of_open_complaints'] = ca_df['number_of_open_complaints'].astype('int')
```

```python
ca_df.info()
```

## Filtering data and Correcting typos

```python
# We need to standardize the text in the column "gender"
ca_df['gender'].value_counts()
```

```python
ca_df['gender'] = ca_df['gender'].fillna('F')
```

```python
# number of missing values in the column "gender"
len(ca_df[ca_df['gender'].isna()==True])
```

```python
# we create a function to clean the "gender column"
def clean_gender(x):
    if x in ['M', 'MALE']:
        return 'Male'
    elif x in["F","female","femal"]:
        return 'Female'
    else:
        return 'U'
```

```python
ca_df['gender'] = list(map(clean_gender, ca_df['gender'])) 
```

```python
#We check we only have 3 categories: Female, male and U
ca_df.gender.unique()
```

```python
#we have to do the same for the columns "state"
ca_df['state'].value_counts()
```

```python
len(ca_df[ca_df['state'].isna()==True])
```

```python
def clean_state(x):
    if x in ['California', 'Cali']:
        return 'california'
    elif x in["Arizona","az"]:
        return 'arizona'
    elif x in ["Washington","wa"]:
         return "washington"
    elif x in ["Oregon"]:
         return "oregon"
    elif x in ["Nevada"]:
        return "nevada"
    else:
        return 'U'
```

```python
ca_df['state'] = list(map(clean_state, ca_df['state'])) 
```

```python
# We only have 5 categories in "state": washington, arizona, nevada, california and oregon
ca_df.state.unique()
```

## Removing duplicates

```python
# we have to use the function drop_duplicates()
ca_df.drop_duplicates()
```

## Replace null values

```python
# in numerical columns, we will replace the NaN values with the mean of each column
# have a loock to our numerical columns
ca_df._get_numeric_data()
```

```python
nulls_df = pd.DataFrame(round(ca_df.isna().sum()/len(ca_df),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
# with this function we can calculate the mean for each column and replace the Nan values

mean_customer_lifetime_value= np.mean(ca_df['customer_lifetime_value'])
ca_df['customer_lifetime_value'] = ca_df['customer_lifetime_value'].fillna(mean_customer_lifetime_value)

mean_income= np.mean(ca_df['income'])
ca_df['income'] = ca_df['income'].fillna(mean_income)

mean_monthly_premium_auto= np.mean(ca_df['monthly_premium_auto'])
ca_df['monthly_premium_auto'] = ca_df['monthly_premium_auto'].fillna(mean_monthly_premium_auto)

mean_total_claim_amount= np.mean(ca_df['total_claim_amount'])
ca_df['total_claim_amount'] = ca_df['total_claim_amount'].fillna(mean_total_claim_amount)
```

```python
nulls_df = pd.DataFrame(round(ca_df.isna().sum()/len(ca_df),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
# We still have some missing data for non numerical variables
```

```python
# We still have some missing values in "policy type","education" and "vehicle_class"

ca_df['policy_type'] = ca_df['policy_type'].fillna(0)
ca_df['education'] = ca_df['education'].fillna(0)
ca_df['vehicle_class'] = ca_df['vehicle_class'].fillna(0)

```

```python
nulls_df = pd.DataFrame(round(ca_df.isna().sum()/len(ca_df),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

## Creating groups of data

```python
# we will create a new column called "region". This new column will replace date from column "state" as follows:

# california = West
# Oregon = North
# washington = East
# Arizona & Nevada = Central

# First, I will create a dictionary with the new labels

```

```python
state_labels ={
    "california":"west",
    "Oregon":"north",
    "washington":"east",
    "arizona":"central",
    "nevada":"central"
}
```

```python
# We can use assign function to append a new column called ""

ca_df=ca_df.assign(region=ca_df.state.map(state_labels))
ca_df
```

```python
# I want to move the column "region" next to the column "state".
list(ca_df.columns)
```

```python
ca_df = ca_df[[
 'state',
 'region',
 'gender',
 'customer_lifetime_value',
 'income',
 'monthly_premium_auto',
 'number_of_open_complaints',
 'policy_type',
 'vehicle_class',
 'total_claim_amount',
 'education']]

ca_df
```
