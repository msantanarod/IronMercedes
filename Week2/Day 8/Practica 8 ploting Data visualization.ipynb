```python
# lineal function
x=np.arange(0,11)
x
```

```python
y1= x*2
y1
```

```python
y2=x/2
y2
```

```python
y3=x**2
y3
```

```python
# 1.  we have to define the figure and give size
plt.figure(figsize=(10,8))
# 2 create the subplot
plt.subplot(1,1,1)
#we have to put fontsize as a dictionary
plt.title("My very first matplot chart ever",{"fontsize":18})
#legends
plt.plot(x,y1,label="y1=x*2")
plt.xlabel("x")
plt.ylabel("y1,y2,y3")
plt.plot(x,y2,label="y2=x/2")
plt.plot(x,y3,label="y3=x**2") # the green line is an exponential relation

#to see the legend:
plt.legend()
```

```python
# we try to do it with subplots
fig,ax=plt.subplots(2,2)
ax[1,0].plot(x,y1,c="#ec42f5")
# "c" is the color that I want to use. we go to google and write "google picker" we choose the color that we want
# and copy the HEX number.We 
```

```python
fig,ax=plt.subplots(2,2,figsize=(20,15))
ax[1,0].plot(x,y1,label="y1=x*2",c="#ec42f5")
ax[1,0].set_title("line function")
ax[1,1].bar(x,y2,label="y2=x/2")
ax[0,0].scatter(x,y3,label="y3=x**2")
# with the multiples plots is a bit difficult to customize the chart. For example creating a label. 
#this is why we introduce figsize in the function
ax[1,0].legend()
ax[1,1].legend()
ax[0,0].legend()

# to see the label we have to introduce the label information in the fuction

# we have to save the chart
plt.savefig("my first chart")
# we can see the chart saved in the same folder that the notebook
```

## practice with sns

```python
# we use Titanic pandas data set

titanic_df=sns.load_dataset("titanic")
titanic_df
```

```python
# we want to see a histogram of the survivals
survival=titanic_df.survived.value_counts()
survival
```

```python
fig.clear()
```

```python
fig, ax = plt.subplots(figsize = (6,6))

ax.bar(survival.index, survival, color=('red','green'),label="survived distribution")
#to set the title
ax.set_title('Titanic Survival Rate', fontweight='bold')
plt.xlabel("survided")
plt.ylabel("count")
ax.legend()
```

```python
classes=titanic_df["class"].value_counts().sort_index()
classes
```

```python
who = titanic_df['who'].value_counts()
who
```

```python
# create figure and axes
fig3, axes3 = plt.subplots(2,1,figsize = (6,12))

# plot the bar chart
axes3[0].bar(who.index, who, color=('red','green', 'blue'))
axes3[1].bar(classes.index, classes, color=('red','green', 'blue'))
```

```python
class_survival = titanic_df.groupby(['class','alive'])['survived'].count().reset_index()
class_survival
```

```python
sns.set_palette("RdYlGn")
fig,ax=plt.subplots(figsize=(6,6))
sns.barplot(x=class_survival["class"],y=class_survival["survived"],hue=class_survival["alive"])
ax.set_title("survival rate by class and alive")
```

```python
#boxplot
fg7 = sns.catplot(x='sex', y = 'age', data = titanic_df, col='class', row='alone', kind="box", margin_titles = True)
fg7.fig.suptitle('Age distribution by class and sex', size='20', fontweight='bold', y = 1.05)
fg7.set_xticklabels(['Male', 'Female'])
fg7.set_ylabels('Age')
fg7.set_xlabels('')
```
