```python
left = pd.DataFrame(
       {
            "key1": ["K0", "K0", "K1", "K2"],
            "key2": ["K0", "K1", "K0", "K1"],
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
        }
    )
    

right = pd.DataFrame(
        {
            "key1": ["K0", "K1", "K1", "K2"],
            "key2": ["K0", "K0", "K0", "K0"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],
        }
    )

print("left \n",left)
print("\nright\n",right)
print("\npd.merge(left, right,how=right, on=[key1, key2],indicator=True)")
result = pd.merge(left, right,how="right", on=["key1", "key2"],indicator=True)
result
```

```python
left = pd.DataFrame(
       {
            "key1": ["K0", "K0", "K1", "K2"],
            "key2": ["K0", "K1", "K0", "K1"],
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
        }
    )
    

right = pd.DataFrame(
        {
            "key1": ["K0", "K1", "K1", "K2"],
            "key2": ["K0", "K0", "K0", "K0"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],
        }
    )

print("left \n",left)
print("\nright\n",right)
print("\npd.merge(left, right,how=right, on=[key1, key2],indicator=True)")
result = pd.merge(left, right,how="inner", on=["key1", "key2"],indicator=True)  #HOW NOS INDICA EL TIPO DE JOIN QUE TENEMOS
result
```
